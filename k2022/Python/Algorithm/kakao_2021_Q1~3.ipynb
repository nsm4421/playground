> 1. 신규 아이디 추천

```python
def solution(new_id):
   
    # step 1
    answer = new_id.lower()

    # step 2
    _list =  [letter for letter in answer  if letter.islower() or letter.isnumeric() or letter in ['-','_','.']]
    answer = ''.join(_list)

    # step 3
    _list = [chunk for chunk in answer.split('.') if chunk!='']
    answer = ''.join([f"{'' if i==0 else '.'}{chunk}" for i, chunk in enumerate(_list)])

    # step 4
    if answer == '.':
        answer = ''
    if answer!='':
        if answer[0] == '.':
            answer=answer[1:]
        elif answer[-1]=='.':
            answer=answer[-1:]

    # step 5
    if answer == '':answer="a"

    # step 6
    if len(answer)>=16:
        answer = answer[:15]
        if answer[-1] == '.':answer=answer[:-1]
   
    # step 7
    if len(answer)<3:
        answer = answer + answer[-1]*(3-len(answer))
  
    return answer
```

```python
import re

def solution(new_id):
    st = new_id
    # 1단계 new_id의 모든 대문자를 대응되는 소문자로 치환합니다.
    st = st.lower()
    # 2단계 new_id에서 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
    st = re.sub('[^a-z0-9\-_.]', '', st)
    # 3단계 new_id에서 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
    st = re.sub('\.+', '.', st)
    # 4단계 new_id에서 마침표(.)가 처음이나 끝에 위치한다면 제거합니다.
    st = re.sub('^[.]|[.]$', '', st)
    # 5단계 new_id가 빈 문자열이라면, new_id에 "a"를 대입합니다.
    # 6단계 new_id의 길이가 16자 이상이면, new_id의 첫 15개의 문자를 제외한 나머지 문자들을 모두 제거합니다.    만약 제거 후 마침표(.)가 new_id의 끝에 위치한다면 끝에 위치한 마침표(.) 문자를 제거합니다.
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub('^[.]|[.]$', '', st)
    # 7단계 new_id의 길이가 2자 이하라면, new_id의 마지막 문자를 new_id의 길이가 3이 될 때까지 반복해서 끝에 붙입니다.
    st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])
    return st
```

> 2. 코스요리

```python
from itertools import combinations
from collections import Counter

def solution(orders:list, course:list)->list:    
    answer = []

    for num_food in course:
        count_list = []
        for order in orders:
            for comb in combinations(order, num_food):
                count_list.append(''.join(sorted(list(comb))))
        candidate = Counter(count_list).most_common()
        answer += [k for k, v in candidate if v>1 and candidate[0][1] == v]      

    return sorted(answer)
```

>3. 순위 검색

```python
def check(info:list, query:list):
    for i, q in zip(info[:-1], query[:-1]):
        if q=='-':
            continue
        elif i != q:
            return False
    if query[-1]=='-':
        return True
    return info[-1]>=query[-1]
            
def preprocess(string):
    li = [word.strip() for word in string.split(' ') if (word !='') and (word!='and')]
    li[-1] = int(li[-1])
    return li

def solution(infos, querys):
    infos = [preprocess(i) for i in infos]
    querys = [preprocess(q) for q in querys]
    result = []
    for query in querys:
        cnt = 0
        for info in infos:
            if check(info, query):
                cnt+=1
        result.append(cnt)
    return result
```

```python
from collections import defaultdict
from itertools import combinations


def solution(infos:list, querys:list):

    answer = []

    info_dict = defaultdict(list)
    for info in infos:
        info = info.split()                       
        score = int(info[-1])                                           # 코딩테스트 점수
        for i in range(5):
            combination = list(combinations(info[:-1], i))
            for c in combination:
                possible_demand_spec = ''.join(c)      
                info_dict[possible_demand_spec].append(score)         

    for spec in info_dict.keys():
        info_dict[spec].sort()                                          # 코딩테스트 점수로 정렬

    for query in querys:
        query_list = query.split()
        for _ in range(3):
            query_list.remove('and')
        while '-' in query_list:
            query_list.remove('-') 

        demand_spec = ''.join(query_list[:-1])                          # 요구 스펙
        cut_line = 0 if query_list[-1] == '-' else int(query_list[-1])   # 코딩테스트 커트라인

        if demand_spec in info_dict.keys():
            candidate_scores = info_dict[demand_spec]  
            # 이진탐색 
            start, end = 0, len(candidate_scores)
            while start<end:
                mid = (start+end)//2
                if candidate_scores[mid]>=cut_line:
                    end = mid
                else:
                    start += 1            
            num_passed = len(candidate_scores) - start 
        else:
            num_passed = 0 

        answer.append(num_passed)
        
    return answer
```

```python
from itertools import combinations
from collections import defaultdict

def solution(infos, querys):
    answer = []
    info_dict = defaultdict(list)
    for info in infos:
        _info = info.split(" ")
        candidate_spec = _info[0:-1]    # 지원자 스펙                                  
        score = int(_info[-1])          # 지원자 코딩테스트 점수

        for r in range(5):
            combination = list(combinations(candidate_spec, r))
            for c in combination:
                possible_demand = ''.join(c)        # 지원자가 통과할 수 있는 요구 스펙
                info_dict[possible_demand].append(score)

    for spec in info_dict.keys():
        info_dict[spec].sort()          # 코딩테스트 점수순으로 정렬

    for query in querys:
        query = query.split()
        cut_line = int(query[-1])       # 코딩테스트 점수 커트라인
        demaned_spec = query[:-1]       # 요구 스펙
        demaned_spec = ''.join(demaned_spec).replace('and','').replace('-','')

        if demaned_spec in info_dict.keys():
            score_list = info_dict[demaned_spec]
            if score_list:
                # 이진탐색
                s, e = 0, len(score_list)
                while s < e:
                    m = (s + e) // 2
                    if score_list[m] >= cut_line:
                        e = m
                    else:
                        s = m+1
                num_pass = len(score_list)-s            # 통과한 사람 수
        else:
            num_pass = 0
        
        answer.append(num_pass)
    return answer
```
